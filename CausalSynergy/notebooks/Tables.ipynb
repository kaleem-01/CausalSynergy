```python
from math import comb

def num_directed_graphs(n):
    return 3 ** (n*(n-1)//2)

def num_dags(n_max):
    a = [1]  # a[0] = 1
    for n in range(1, n_max+1):
        total = 0
        for k in range(1, n+1):
            total += (-1)**(k-1) * comb(n,k) * (2 ** (k*(n-k))) * a[n-k]
        a.append(total)
    return a

# Example for n=2..8
n_max = 8
a = num_dags(n_max)
for n in range(2, n_max+1):
    tot = num_directed_graphs(n)
    dags = a[n]
    pct = dags / tot * 100
    print(f"n={n}: total={tot:,}, DAGs={dags:,}, percent={pct:.2f}%")
```
