```python
import networkx as nx

import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_nodes_from(['A', 'B', 'C'])

# Add edges to form a collider structure: A -> C <- B
G.add_edge('A', 'C')
G.add_edge('B', 'C')

# Draw the graph
# pos = {'A': (-1, 0), 'B': (1, 0), 'C': (0, -1)}
# Move nodes closer to C to shorten edge length
pos = {'A': (-0.05, 0), 'B': (0.05, 0), 'C': (0, -0.15)}
nx.draw(
    G,
    pos,
    with_labels=True,
    arrows=True,
    node_color='white',
    edgecolors='black',
    node_size=2500,
    font_size=20,
    width=2,
    arrowstyle='-|>',
    arrowsize=20,
    font_family='Times New Roman',
)
# nx.draw(G, pos, with_labels=True, arrows=True, node_color='lightblue', node_size=1500, font_size=16)
# plt.title("Collider Structure: A → C ← B")
plt.gcf().set_size_inches(4, 4)
# plt.tight_layout()
plt.savefig("../plots/Illustrations/ColliderStructure.png", dpi=800)
```
