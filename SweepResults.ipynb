```python
import pandas as pd
```

```python
results_er = pd.read_csv("results/param_sweeps/sweep_trials_er.csv")
results_ws = pd.read_csv("results/param_sweeps/sweep_trials_ws.csv")
results_ba = pd.read_csv("results/param_sweeps/sweep_trials_ba.csv")
```

```python
for name, results in [("ER", results_er), ("WS", results_ws), ("BA", results_ba)]:
    print(f"Results {name}:")
    print("Total number of trials:", len(results))
    
    sorted_results = results.sort_values(by="best_score", ascending=False)
    display(sorted_results.head(10))  # change/remove head(10) to show more
```
