#### Two Roots

```python
configs = {
        'folder': 'data/precomputed_cpds/two_roots',
        'algorithm': 'genetic-synergy',
        'num_vars': 20,
        'num_samples': 10_000,
        'replications': 1
}

pipeline_tworoots = Pipeline(configs, run_ea_synergy)
pipeline_tworoots.get_true_dags()
pipeline_tworoots.get_results()
pipeline_tworoots.get_metrics()
```

```python

from src.graphviz import GraphVisualization
from pgmpy.estimators import BIC

idx = 0

metadata = dill.load(open(f"{pipeline_tworoots.configs["folder"]}/{pipeline_tworoots.metadata_by_id[idx]}", "rb"))
df = pd.read_csv(f"{pipeline_tworoots.configs['folder']}/{pipeline_tworoots.csv_by_id[idx]}")


graph_viz = GraphVisualization(pipeline_tworoots.true_dags[idx], pipeline_tworoots.learned_dags[idx],
                               metadata=metadata)
graph_viz.compare_dags_with_labels()

scorer = BIC(df)
mapping = {n: str(n) for n in pipeline_tworoots.true_dags[idx].nodes()}  # convert all int nodes to str
G = nx.relabel_nodes(pipeline_tworoots.true_dags[idx], mapping)
print("BIC score for the true DAG:", scorer.score(G))
scorer.score(G)

print("BIC score for the learned DAG:", scorer.score(pipeline_tworoots.learned_dags[idx]))
print(metadata)
print(pipeline_tworoots.csv_by_id[idx])
print(pipeline_tworoots.metadata_by_id[idx])
```
